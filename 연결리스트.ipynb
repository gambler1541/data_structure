```python
class Slist:
    class Node:
        def __init__(self, item, link):
            self.item = item
            self.next = link
            
    def __init__(self):
        self.head = None
        self.size = 0
        
    def size(self):
        return self.size
    
    def is_empty(self):
        if self.size == 0:
            return True
        
    def insert_front(self, item):
        # is_empty가 True일 경우
        if self.is_empty():
            # 새 노드를 head값으로 생성
            self.head = self.Node(item, None)
        else:
            # 노드가 있을경우
            # head가 새 노드를 가리킴
            # 기존 head노드와 연결
            self.head = self.Node(item, self.head)
        # 노드를 생성하고 size를 1증가 
        self.size += 1
        
    def insert_after(self, item, p):
        # 데이터의 가장 마지막에 노드를 삽입
        # item은 새 노드의 데이터
        # 새 노드가 p의 다음 노드가 됨
        p.next = Slist.Node(item, p.next)
        self.size += 1
    
    def delete_front(self):
        # head노드를 삭제할 경우
        if self.is_empty():
            raise EmptyError('Underflow')
        else:
            # head의 다음 값을 head로 참조
            self.head = self.head.next
            self.size -= 1
    
    def delete_after(self, p):
        # p의 다음노드를 삭제 할 경우
        if self.is_empty():
            raise EmptyError('Underflow')
        # p의 다음 값을 t에 저장
        t = p.next
        # p의 다음 값을 t의 다음 값으로 저장 후
        # 삭제하고자 하는 t는 함수의 호출이 끝나면 레퍼런스 카운트가 0이되면서 사라짐
        p.next = t.next
        self.size -= 1
        
    def search(self, target):
        # target을 찾는 메서드
        # p에 head값을 참조(Linked_list의 시작)
        p = self.head
        # 순회
        #  size의 길이만큼 순회를 함
        for k in range(self.size):
            # target과 p의 data가 같다면
            if target == p.item:
                # p가 있는 index값을 리턴
                return k
            # p의 data가 target과 다르다면 p의 다음값을 참조
            p = p.next
        # 찾는 target이 존재하지 않는다면 => None
        return None
        
    def print_list(self):
        p = self.head
        while p:
            if p.next != None:
                print(p.item, '->', end=' ')
            else:
                print(p.item)
            p = p.next
class EmptyError(Exception):
    pass
```

```python
s = Slist()
s.insert_front('oragne')
s.insert_front('apple')
s.insert_after('cherry', s.head.next)
s.insert_front('pear')
s.print_list()
print(f'cherry는 {s.search("cherry")}')
print(f'kiwi는 {s.search("kiwi")}')
print('배 다음 노드 삭제 후:', end='')
s.delete_after(s.head)
s.print_list()
print('첫 노드 삭제 후:', end='')
s.delete_front()
s.print_list()
print('첫 노드로 망고, 딸기 삽입 후:', end='')
s.insert_front('mango')
s.insert_front('strawberry')
s.print_list()
print('오렌지 다음 노드 삭제 후:', end='')
s.delete_after(s.head.next.next)
s.print_list()
```
